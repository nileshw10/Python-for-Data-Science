**Aim**

To perform data wrangling operations using Pandas such as filtering data, grouping, merging datasets, handling missing values, and applying transformations to prepare data for analysis.

**Theory**
1. What is Data Wrangling?

Data wrangling is the process of cleaning, organizing, and transforming raw data into a useful format for analysis. It includes handling missing values, correcting errors, and restructuring data. Data preprocessing is a part of data wrangling that mainly focuses on preparing data for machine learning models, while data wrangling is a broader process that includes cleaning and transforming data for general use.

2. Importance of EDA in Pandas

Exploratory Data Analysis (EDA) helps us understand the dataset before applying any model. Using Pandas, we can easily check data structure, detect missing values, find patterns, and summarize data. It helps in making better decisions and improving data quality.

3. Difference between Series, DataFrame, and Panel

A Series is a one-dimensional array that holds data like a single column. A DataFrame is a two-dimensional table with rows and columns. A Panel is a three-dimensional data structure, but it is now removed in newer versions of Pandas and replaced by other methods.

4. Functions in Pandas

a. isnull()
It is used to check missing values in the dataset and returns True if the value is missing.

b. notnull()
It returns True for non-missing values and False for missing values.

c. dropna()
It removes rows or columns that contain missing values.

d. fillna()
It fills missing values with a specified value like mean, median, or any constant.

5. Column Transformation

Column transformation means modifying the values of a column using operations or functions. It is important because it helps in converting data into a suitable format for analysis, like scaling values or creating new features.

```python
import pandas as pd
df = pd.read_csv("/content/employee updated (1).csv")
print(df)

# 1. Select a single column from a DataFrame
name = df['Name']
print(name.head(5))

# 2. Add a new column with constant values
df['Constant_Value'] = 10
print(df.head())

# 3. Sort a DataFrame by one column
df_sorted_age = df.sort_values('Age')
print(df_sorted_age.head(5))

# 4. Rename DataFrame columns
df_renamed = df.rename(columns={'Employee_ID': 'ID'})
print(df_renamed.head())

# 5. Display summary statistics using describe()
print(df.describe())

# 6. Remove duplicate rows from a dataset
df_no_duplicates = df.drop_duplicates()
print(df_no_duplicates.head(10))
display(df_renamed.head())

# 7. Handle missing values using mean or median
print("Handling missing data")
df.loc[2, 'Salary'] = None  # Introducing NaN (missing value) in the 'Score' column
print(df)
df['Salary'] = df['Salary'].fillna(df['Salary'].mean())
print("\nAfter filling missing 'Score' values with mean:")
print(df)

# First, let's re-introduce a missing value in 'Bonus' for demonstration
df.loc[0, 'Bonus'] = None

print("Missing values before handling with mean and median:")
print(df[['Salary', 'Bonus']].isnull().sum())

# Fill missing 'Salary' values with the mean
df['Salary'] = df['Salary'].fillna(df['Salary'].mean())

# Fill missing 'Bonus' values with the median
df['Bonus'] = df['Bonus'].fillna(df['Bonus'].median())

print("\nMissing values after handling with mean and median:")
print(df[['Salary', 'Bonus']].isnull().sum())

print(df.head())

# 8. Change data types of selected columns
print("Data types before change:\n", df.dtypes)
# Convert 'Join_Date' to datetime type
df['Join_Date'] = pd.to_datetime(df['Join_Date'])
# Convert 'Gender', 'Department', and 'Experience/Fresher' to categorical types
if 'Gender' in df.columns:
    df['Gender'] = df['Gender'].astype('category')
if 'Department' in df.columns:
    df['Department'] = df['Department'].astype('category')
if 'Experience/Fresher' in df.columns:
    df['Experience/Fresher'] = df['Experience/Fresher'].astype('category')

print("\nData types after change:\n", df.dtypes)
print(df.head())

# 9. Filter rows based on a given condition
# Filter employees older than 30 years
df_filtered_age = df[df['Age'] > 30]
print(f"Number of employees older than 30: {len(df_filtered_age)}")
print(df_filtered_age.head())

# 10. Save the cleaned dataset to a new CSV file
df.to_csv('employee cleaned.csv', index=False)
print(" Data saved to 'employes cleaned .csv'")
```
